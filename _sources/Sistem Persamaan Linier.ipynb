## **Sistem Persamaan Linier**


Sistem persamaan linier juga bisa disebut dengan persamaan aljabar. karakteristik dari SPL ini memiliki setiap sukunya mengandung konstanta atau perkalian konstanta dengan variabel tunggal. Dalam persamaan linier akan ada variabel, koefisien dan konstanta. Variabel adalah simbol pengganti nilai numerik, koefisien merupakan bilangan yang akan menyatakan jumlah suatu variabel.


## ciri-ciri persamaan linier
1. persamaan linier akan memilki pangkat satu
2. persamaan linier tidak memiliki perkalian variabel
3. biasanya terdiri dari dua ruas yang akn dihubungkan dengan tanda sama dengan
4. perkalian, penjumlahan, pengurangan, dan pembagian pada kedua ruas persamaan linier tidak akan mengubah nilai persamaan.


## Unsur-unsur dalam persamaan linier
1. variabel adalah penambah atau pengganti bilangan yang nilainya belum diketahui besarannya. contoh: a,b,c,x,y,z
2. koefisien adalah bilangan yang menjelaskan banyaknya jumlah data
3. Konstanta adalah nilai yang konstan yang tidak diikuti oleh variabel dibelakangnya.

## Jenis-jenis persamaan linier

1. persamaan linier satu variabel
merupakan persamaan yang memiliki satu variabel dengan pangkat satu. contoh: 
ax + b = 0

a = koefisien

b = konstanta

x = variabel

2. persamaan linier dua variabel
adalah yang memiliki dua variabel untuk menggunakan relasi = dan tidak akan menggunakan perkalian variabel pada setiap persamaan. contoh : ax + by = 0
3. persamaan linier tiga variabel
persamaan ini bisa dilakukan dengan dua cara menggunakan metode subsitusi dan metode eliminasi gauss. contoh: 
ax + by +cy = d

## **Metode Penyelesaian persamaan linier**

## **1. Metode subsitusi**

ini merupakan salah satu cara mengubah suatu variabel dengan variabel persamaan lainnya. contoh
$2x + 4y$

$2x + 4y = 28$
selanjutnya pindahkan sisi ruang kanan
$2x = 28 -4y$
memilih variabel y untuk dipindahkan maka koefisien pada variabel x akan dihilangkan dengan cara membagi dengan niali koefisien dari x.
$2x/2 = 28 -4y/2$
dan ini menghasilkan persamaan x = 14 -2y  sebagi bentuk solusi.
selanjutnya bisa menggabungkan persamaan $3x + 2y = 22$ dengan $x = 14 -2y$ dimana ini akan mengganti variabel x dengan persamaan.
x = 14-2y

## **2. Metode Eliminasi Gauss**
metode ini digunakan untuk menyelesaikan soal persamaan linier dua atau tiga variabel dengan mengubah menjadi bentuk matriks, lalu akan diubah ke bentuk eselon baris melalui baris elementer.

## Contoh Penyelesaian Sistem Linier

```python
import numpy as np
A=np.array([[8,9,1,19],[1,2,3,9],[2,1,1,5]])
print(A)
```

```python
def RowSwap(A,k,l):
    # =============================================================================
#     A is a NumPy array.  RowSwap will return duplicate array with rows
#     k and l swapped.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp

    return B

def RowScale(A,k,scale):
# =============================================================================
#     A is a NumPy array.  RowScale will return duplicate array with the
#     entries of row k multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale

    return B

def RowAdd(A,k,l,scale):
# =============================================================================
#     A is a numpy array.  RowAdd will return duplicate array with row
#     l modifed.  The new values will be the old values of row l added to
#     the values of row k, multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[l][j] += B[k][j]*scale

    return B
```

```python
# menambah baris 0 dengan baris 1 yang sudah dikalikan -2
A1 = RowAdd (A,0,1,-2)
print(A1)
```

```python
# Menukar baris 0 dan 1
A1 = RowSwap(A,0,1)
print(A1, '\n')
```

```python
# Menambahkan baris 0 ke baris 1 yang sudah dikalikan -8
A2 = RowAdd(A1,0,1,-8)
print(A2,'\n')
```

```python
# menambahkan baris 0 ke baris 2 yang sudah dikalikan -2
A3 = RowAdd(A2,0,2,-2)
print(A3,'\n')
```

```python
# mengalikan baris 1 dengan -1.0/7
A4 = RowScale(A3,1,-1.0/7)
print(A4,'\n')
```

```python
# menambahkan baris 1 ke baris 2 yang sudah dikalikan 3
A5 = RowAdd(A4,1,2,3)
print(A5,'\n')
```

```python
# mengalikan baris 2 dengan 1/4.85714286
A6 = RowScale(A5,2,1/4.85714286)
print(A6,'\n')
```
